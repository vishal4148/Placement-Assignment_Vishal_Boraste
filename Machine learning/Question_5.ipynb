#### Q-5. Uber is a taxi service provider as we know, we need to predict the high booking area using an Unsupervised algorithm and price for the location using a supervised algorithm and use some map function to display the data

```python
import pandas as pd
df = pd.read_csv("rideshare_kaggle.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df.drop(columns=['timezone', 'datetime', 'product_id', 'latitude', 'longitude', 'id'])
df.shape
```

```python
df.info()
```

```python
df_categorical = df[['source', 'destination', 'cab_type','name','short_summary','long_summary']]
df_categorical.describe()
```

```python
duplicate = df.duplicated()
print(df[duplicate])
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
category_f = ['source', 'destination', 'cab_type','name','short_summary','long_summary','icon']
from sklearn.preprocessing import LabelEncoder

for data in category_f:
    encoder = LabelEncoder()
    df[data] = encoder.fit_transform(df[data])
```

```python
df.info()
```

```python
X = df.drop(columns = "price")
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
a = LinearRegression()
a.fit(X_train,y_train)
a_pred = a.predict(X_test)
mse = mean_squared_error(y_test, a_pred)
print(mse)
```
