#### Write a program to download the data from the given API link and then extract the following data with proper formatting

Link - http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes 

Note - Write proper code comments wherever needed for the code understanding



Excepted Output Data Attributes -

● id - int url - string

● name - string season

● - int number - int

● type - string airdate -

● date format airtime -

● 12-hour time format

● runtime - float

● average rating - float

● summary - string

● without html tags

● medium image link - string

● Original image link - string

```python
import requests

class TVShowData:
    def __init__(self, show_name):
        self.show_name = show_name
        self.api_url = f"http://api.tvmaze.com/singlesearch/shows?q={show_name}&embed=episodes"
    
    def download_data(self):
        # Download data from the API
        response = requests.get(self.api_url)
        self.data = response.json()
    
    def extract_attributes(self):
        # Extract the required data attributes
        self.show_id = self.data.get("id", "")
        self.show_url = self.data.get("url", "")
        self.show_name = self.data.get("name", "")
        self.season = self.data.get("season", "")
        self.episode_number = self.data.get("number", "")
        self.show_type = self.data.get("type", "")
        self.airdate = self.data.get("airdate", "")
        self.airtime = self.data.get("airtime", "")
        self.runtime = self.data.get("runtime", "")
        self.average_rating = self.data.get("rating", {}).get("average", "")
        self.summary = self.data.get("summary", "").replace("<p>", "").replace("</p>", "")
        self.image = self.data.get("image", {})
        self.medium_image_link = self.image.get("medium", "")
        self.original_image_link = self.image.get("original", "")
    
    def print_attributes(self):
        # Print the extracted data attributes
        print("ID:", self.show_id)
        print("URL:", self.show_url)
        print("name:", self.show_name)
        print("season:", self.season)
        print("number:", self.episode_number)
        print("type:", self.show_type)
        print("airdate:", self.airdate)
        print("airtime:", self.airtime)
        print("runtime:", self.runtime)
        print("average rating:", self.average_rating)
        print("summary:", self.summary)
        print("medium image link:", self.medium_image_link)
        print("original image link:", self.original_image_link)

# Create an instance of the TVShowData class
tv_show = TVShowData("westworld")

# Download and extract data
tv_show.download_data()
tv_show.extract_attributes()

# Print the extracted data attributes
tv_show.print_attributes()
```
